```python
import numpy as np
from PIL import Image as im
from IPython.display import display
```

```python
image_open = im.open("rana.jpg")
display(image_open)
```

```python
# Image Details
print(image_open.format)
print(image_open.size)
print(image_open.mode)
```

### convert Image pixel into numpy array

```python
array = np.array(image_open)
```

```python
array
```

```python
array.shape
```

### Create another 3d Array with the value of 0 to 255 here 0 is for black color and 255 is for white color

```python
array2 = np.full(array.shape, 255)
```

```python
array2
```

### Now Let's Subtract array from initial array to second created array

```python
new_arr = array-array2
```

```python
display(new_arr)
```

### Now convert negative values into positive values

```python
new_arr = new_arr*-1
```

### Now Let's define array data type

```python
new_arr = new_arr.astype(np.uint8)
```

```python
display(im.fromarray(new_arr))
```
